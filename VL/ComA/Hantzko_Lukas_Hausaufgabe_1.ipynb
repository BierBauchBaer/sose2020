### a) Ausgabeformatierung

```python
f(x) = x^2/(2*(x-1)^2)

plot(f)
```

```python
def rootStr(func):
    values = func.roots()
    Ausgabe = "Nullstelle\tVielfachheit\n"
    for i in values:
        Ausgabe += str(n(i[0],digits=5))+"\t\t"+str(i[1])
        Ausgabe += "\n"
    return Ausgabe
        
print rootStr(f)
```

### b) Polstellen

```python
def polStr(func):
    invFunc = 1/func
    values = invFunc.roots()
    Ausgabe = "Polstelle\tVielfachheit\n"
    for i in values:
        Ausgabe += str(n(i[0],digits=5))+"\t\t"+str(i[1])
        Ausgabe += "\n"
    return Ausgabe

print polStr(f)
```

### c) Extremstellen

```python
def extrStr(func):
    derFunc = func.derivative(x)
    secFunc = derFunc.derivative(x)
    values = derFunc.roots()
    Ausgabe = "Extremstelle\tArt\n"
    for i in values:
        Ausgabe += str(n(i[0],digits=10))+"\t"
        secDer = secFunc(i[0])
        if secDer == 0:
            Ausgabe += "Sattelpunkt"
        elif secDer > 0:
            Ausgabe += "Minimum"
        else:
            Ausgabe += "Maximum"
        Ausgabe += "\n"
    return Ausgabe

print extrStr(f)
```

### d) Asymptoten

```python
if Abgabe:
    help(limit)
    
def asymptotes(func):
    Asymptoten = []
    for i in (-1,1):
        m = limit(func/x,x=i*Infinity)
        n = limit(func-m*x,x=i*Infinity)
        g(x) = m*x + n
        Asymptoten.append(g)
    return Asymptoten

print asymptotes(f)
```

### e) Plot

```python
plot(asymptotes(f),color=["green","red"])+plot(f,detect_poles=True)
```
